```python
def plot_potential(K1=1, K2=1, C=0, n_contours=10, n_points=50):
    #define mesh and meshgrid for plotting
    q1 = np.linspace(-1,1,n_points)
    q2 = q1
    Q1, Q2 = np.meshgrid(q1, q2)
    Z = potential_energy(Q1, Q2, K1, K2, C)

    #construct contour plot
    plt.figure(figsize=(3,2))
    plt.contourf(Q1, Q2, Z, levels=n_contours)
    plt.colorbar(label='Potential Energy')
    plt.xlim([-1,1])
    plt.ylim([-1,1])
    plt.xlabel('$Q_1$')
    plt.ylabel('$Q_2$')

    plt.show()

plot_potential(K1=1, K2=1, C=0, n_contours=10, n_points=50)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

def potential_energy(Q1, Q2, K1=1.0, K2=1.0, C=0):
    return 0.5 * K1 * Q1**2 + 0.5 * K2 * Q2**2 + C * Q1 * Q2

q1 = np.linspace(-1,1,3)
q2 = q1
Q1, Q2 = np.meshgrid(q1, q2)

potential_energy(Q1, Q2)

plt.figure(figsize=(3,2))
plt.contourf(Q1, Q2, potential_energy(Q1, Q2, K1=.5, K2=1, C=-.7), levels=25)
plt.colorbar(label='Potential Energy')
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.xlabel('$Q_1$')
plt.ylabel('$Q_2$')

def plot_potential(K1=1, K2=1, C=0, n_contours=10, n_points=50):
    #define mesh and meshgrid for plotting
    q1 = np.linspace(-1,1,n_points)
    q2 = q1
    Q1, Q2 = np.meshgrid(q1, q2)
    Z = potential_energy(Q1, Q2, K1, K2, C)

    #construct contour plot
    plt.figure(figsize=(3,2))
    plt.contourf(Q1, Q2, Z, levels=n_contours)
    plt.colorbar(label='Potential Energy')
    plt.xlim([-1,1])
    plt.ylim([-1,1])
    plt.xlabel('$Q_1$')
    plt.ylabel('$Q_2$')

plot_potential(K1=1, K2=1, C=0, n_contours=10, n_points=50)

widgets.interact(plot_potential,
                K1=widgets.FloatSlider(value=1.0, min=0.0, max=2.0, step=0.1, description='K1'),
                K2=widgets.FloatSlider(value=1.0, min=0.0, max=2.0, step=0.1, description='K2'),
                C=widgets.FloatSlider(value=1.0, min=-1.0, max=1.0, step=0.05, description='C'),
       
                n_contours=widgets.IntSlider(value=1.0, min=1.0, max=50.0, step=1, description='n_contours'),
                n_points=widgets.IntSlider(value=1.0, min=10.0, max=60.0, step=1, description='n_points'),
                )
```

```python
1. As you increase $C$, how do the low-energy contours start to change?
2. Does the force on $Q_1$ depend on the direction of $Q_2$?
3. What do you expect the influence of driving $Q_2$ to be on the motion of $Q_1$?
4. If $Q_1$ is kicked away from equilibrium, what do you expect its influence to be on $Q_2$?

1. As you increase C the low energy contours began to completely flip to the opposite side they started at. Meaning the low-energy contours will
began to shift towards Q1 and Q2 which shows the potential energy intertwining with itself.

2. Yes the force acting on Q1 is dependent on the value of Q2. (C * Q1 * Q2) This portion of the potential energy function directly intels
the relationship between the two coupled oscillators. This also means if one has a different sign(+ or -) it will have a direct impact.

3. There would be some sort of kinetic energy transfered into Q1 which could knock it out of equilibrium. No matter what the two are coupled together
meaning there is bound to be an effect on the system as a whole.
                                                                                        
4. If Q1 is knocked out of equilibrium, the force with cause a change in Q2. However this force will depend on the systems parameters and how strong
the Qs are coupled together. 
```
