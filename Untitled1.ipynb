```python
##Our Plan
1. Import Numoy, Matplotlib, ipywidegts
2. Define the potential energy
3. Define meshes for $Q_1$ and $Q_2$
4. use ipywidgets to create interactive plots(contours in our case)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
def potential_energy(Q1, Q2, K1=1.0, K2=1.0, C=0):
    return 0.5*K1*Q1**2 + 0.5*K2*Q2**2 + C*Q1*Q2**2
potential_energy(1,1)
```

```python
#define meshes in Q1 and Q2
import numpy as np
q1 = np.linspace(-1,1,3)
q2 = q1
Q1, Q2 = np.meshgrid(q1, q2)

potential_energy(Q1, Q2)
```

```python
#plot contours with matplotlib contour
import matplotlib.pyplot as plt
plt.figure(figsize=(3,2))
plt.contourf(Q1, Q2, potential_energy(Q1, Q2, K1=.5, K2=1, C=-.7), levels=25)
plt.colorbar(label='Potential Energy')
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.xlabel('$Q_1$')
plt.ylabel('$Q_2$')

plt.show
```

```python
#define plotting function
def plot_potential(K1=1, K2=1, C=0, n_contours=10, n_points=50):
    #define mesh and meshgrid for plotting
    q1 = np.linspace(-1,1,n_points)
    q2 = q1
    Q1, Q2 = np.meshgrid(q1, q2)
    Z = potential_energy(Q1, Q2, K1, K2, C)

    #construct contour plot
    plt.figure(figsize=(3,2))
    plt.contourf(Q1, Q2, Z, levels=n_contours)
    plt.colorbar(label='Potential Energy')
    plt.xlim([-1,1])
    plt.ylim([-1,1])
    plt.xlabel('$Q_1$')
    plt.ylabel('$Q_2$')

    plt.show
    return

plot_potential(K1=1, K2=1, C=0, n_contours=10, n_points=50)
```

```python
import ipywidgets as widgets
#help(widgets.FloatSlider)

widgets.interact(plot_potential,
                K1=widgets.FloatSlider(value=1.0, min=0.0, max=2.0, step=0.1, description='K1'),
                K2=widgets.FloatSlider(value=1.0, min=0.0, max=2.0, step=0.1, description='K2'),
                C=widgets.FloatSlider(value=1.0, min=-1.0, max=1.0, step=0.05, description='C'),
                n_contours=widgets.FloatSlider(value=1.0, min=1.0, max=50.0, step=1, description='n_contours'),
                n_points=widgets.FloatSlider(value=1.0, min=10.0, max=60.0, step=1, description='n_points'),
                )
```
