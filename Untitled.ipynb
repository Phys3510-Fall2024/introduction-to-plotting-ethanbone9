```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define the potential energy
def potential_energy(x, k=1.0):
    return 0.5 * k * x**2

#test function
potential_energy(2, 10)
```

```python
import numpy as np
x = np.linspace(-10, 10, 100)

y = potential_energy(x)

print(np.transpose([x, y]))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,4))
plt.plot(x, y, color="black", linewidth=0.5)
plt.xlabel("displacement (arbitrary unit)")
plt.ylabel("potential energy (arbitrary unit)")

plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
```

```python
#contour plots
import numpy as np
import matplotlib.pyplot as plt
```

```python
def potential_energy_2D(x, y, kx=1.0, ky=1.0):
    return 0.5 * (kx * x**2 + ky * y**2)

#test function
potential_energy_2D(2, 1)
```

```python
x = np.linspace(-5, 5, 3)
y = np.linspace(-5, 5, 3)

X, Y  = np.meshgrid(x, y)

Z = potential_energy_2D(X, Y)

print(X)
print(Y)
print(Z)
```

```python
plt.figure(figsize=(4,4))
plt.contourf(X, Y, Z)

plt.colorbar(label = "potential energy (arbitrary unit)")


plt.xlabel("x-displacement (arbitrary unit)")
plt.ylabel("y-displacement (arbitrary unit)")
plt.xlim(-4, 4)
plt.ylim(-4, 4)

plt.show()
```
